### Как менялись продажи по платформам

Определим по закону Парето платформы, которые дали 80% продаж за все время, для этих платформ посчитаем годы, за которые для платформы были получены 80% продаж по играм по годам.

```python
# Определим Платформы, давшие 80% продаж
sales_pareto = df['total_sales'].sum()*0.8 
sales_series = df.groupby('platform')['total_sales'].sum().sort_values(ascending=False)
pareto_index = 0
try:
    while sales_pareto >= 0:
        sales_pareto -= sales_series[pareto_index]
        pareto_index += 1       
    plat_list = sales_series[0:pareto_index].index.tolist()
except:
        print('ошибка в данных при обработке циклом \'пока\'')
        
# Определим года по этим платформам, давшие 80% продаж
years_success = 0
for val in plat_list:
    len_good_platform = len(plat_list)
    index_years = 0
    sales_platform = df.query('platform == @val')['total_sales'].sum()*0.8
    sales_per_years = df.query('platform == @val').groupby('year_of_release')['total_sales'].sum().sort_values(ascending=False)
    try:
        while sales_platform >=0:
            sales_platform -= sales_per_years[index_years]
            index_years += 1
        years_success += index_years
    except:
        print('ошибка в данных при обработке циклом \'пока\'')

print(f'''Исходя из закона Парето выбрали платформы: {plat_list}
среднее время пика популярности платформы по Парето и появления новой - {round(years_success/len_good_platform, 1)} года''')
      
```

Выведем продажи по всем годам, для опреденных самых популярных платформ:

```python
# Создаем сводную таблицу где строки - годы, столбцы - платформы, значения - суммарные продажи
pivot_subset = df.query('platform in @plat_list').pivot_table(index='year_of_release', columns='platform',values='total_sales',aggfunc='sum')
pivot_subset.index = pd.DatetimeIndex(pivot_subset.index).year

plt.figure(figsize=(14,14))
sns.heatmap(pivot_subset,annot=True,fmt='4.2f', cmap='OrRd')
plt.show()
```

```python
#Создадим функцию вывода графика heatmap
def heatmap_variable(data,tit):
    plt.figure(figsize=(11,6))
    sns.heatmap(data,annot=True,fmt='4.2f', cmap='OrRd').set_title(tit)
    plt.show()
#Создадим функцию сводной таблицы обычной и в процентах от общей суммы столбцов   
def prosent_for_column(data, ind='', p_col='', p_val='', aggf='mean', procent=False):
    if ind == '' or p_col=='' or p_val=='':
        print('Введите значения ind(индекс/сы для сводной, p_col(колонка/и для сводной), p_val(значение/я для сводной)')
    elif procent == True: # Вывод сводной таблицы с подсчетом % от общей суммы колонок
        subset_data = data.pivot_table(index=ind, columns=p_col,values=p_val,aggfunc=aggf).reset_index()
        list_col = subset_data.columns.tolist()[1:13]
        subset_data['tts'] = subset_data.apply((lambda z: z[list_col].sum(axis=0)),axis=1)
        for v in list_col:
            subset_data[v] = subset_data.apply((lambda c: c[v]/c['tts']*100), axis=1)
        subset_data = subset_data.set_index(subset_data.columns.tolist()[0]).drop(columns ='tts', axis = 1)
        subset_data.index = pd.DatetimeIndex(subset_data.index).year
        return subset_data
    else: # Вывод сводной таблицы
        subset_data = data.pivot_table(index=ind, columns=p_col,values=p_val,aggfunc=aggf)
        subset_data.index = pd.DatetimeIndex(subset_data.index).year
        return subset_data
    
```

```python

potencial_platform_data = df.query('year_of_release >= 2010')
subset_potenc = prosent_for_column(potencial_platform_data, 'year_of_release','platform','total_sales','sum')
heatmap_variable(subset_potenc,'platform total sales')

subset_plat_sales = prosent_for_column(potencial_platform_data, 'year_of_release','platform','total_sales','sum', procent=True)
heatmap_variable(subset_plat_sales, 'platform total sales in %')
```

```python
#Создаем функцию обработки данных по регионам через сводную таблицу и с выводом либо head, либо сводной таблицы
def print_top_5_for_region(data, group_col, text, func='sum', head_num=10, pivot=False):
    data_pivot = data.pivot_table(index=group_col, values=['na_sales','eu_sales','jp_sales', 'other_sales'], aggfunc=func)
    if pivot == False:
        reg_0 = data_pivot['na_sales'].sort_values(ascending=False).head(head_num)
        reg_1 = data_pivot['eu_sales'].sort_values(ascending=False).head(head_num)
        reg_2 = data_pivot['jp_sales'].sort_values(ascending=False).head(head_num)
        reg_3 = data_pivot['other_sales'].sort_values(ascending=False).head(head_num)
        
        print(f'''Топ-5 самых популярных {group_col} для регионов продаж за {text}:
        {reg_0}
    
        {reg_1}
    
        {reg_2}
    
        {reg_3}
            ''')
        plt.pie(reg_0,labels=reg_0.index)
        plt.title('na_sales')
        plt.show()
    
        plt.pie(reg_1,labels=reg_1.index)
        plt.title('eu_sales')
        plt.show()
    
        plt.pie(reg_2,labels=reg_2.index)
        plt.title('jp_sales')
        plt.show()
    
        plt.pie(reg_3,labels=reg_3.index)
        plt.title('other_sales')
        plt.show()
    else:
        print(f'Сводная таблица по {group_col} для регионов продаж за {text}:')
        display(data_pivot.sort_values(by='na_sales',ascending=False))
   
```

## Проверьте гипотезы

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
Задайте самостоятельно пороговое значение alpha.
Поясните:
- Как вы сформулировали нулевую и альтернативную гипотезы;
- Какой критерий применили для проверки гипотез и почему.

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Сформулируем 0-гипотезу: ***Средние пользовательские рейтинги платформ Xbox One и PC одинаковые***

Сформулируем альтернативную гипотезу: ***Средние пользовательские рейтинги платформ Xbox One и PC разные***

Проверим с момощью сравнения выборок из двух независимых генеральных совокупностей:

```python
print(f'''Выборка 1 количество значений: {potencial_platform_data.query('platform == "XOne" and user_score <= 10')['user_score'].count()},
Выборка 2 количество значений: {potencial_platform_data.query('platform == "PC" and user_score <= 10')['user_score'].count()},
количество данных в выборках достаточное, чтобы считать дисперсии равными.''')
```

```python
alpha = 0.01

results = st.ttest_ind(
   potencial_platform_data.query('platform == "XOne" and user_score <= 10')['user_score'],
   potencial_platform_data.query('platform == "PC" and user_score <= 10')['user_score']) # ваш код: проверьте гипотезу о равенстве средних выручек независимых совокупностей Москва и остальные города

print('p-значение:', results.pvalue)

if results.pvalue < alpha: # сравниваем получившееся p-значение с заданным уровнем статистической значимости:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Вывод** Гипотеза ***Средние пользовательские рейтинги платформ Xbox One и PC одинаковые*** подтверждена с заданной статичтической значимостью

### Средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем 0-гипотезу: ***Средние пользовательские рейтинги жанров Action и Sports одинаковые.***

Сформулируем альтернативную гипотезу: ***Средние пользовательские рейтинги жанров Action и Sports разные.***

Проверим с момощью сравнения выборок из двух независимых генеральных совокупностей:

```python
alpha = 0.01

results = st.ttest_ind(
   potencial_platform_data.query('genre == "action" and user_score <= 10')['user_score'],
   potencial_platform_data.query('genre == "sports" and user_score <= 10')['user_score']) # ваш код: проверьте гипотезу о равенстве средних выручек независимых совокупностей Москва и остальные города

print('p-значение:', results.pvalue)

if results.pvalue < alpha: # сравниваем получившееся p-значение с заданным уровнем статистической значимости:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
